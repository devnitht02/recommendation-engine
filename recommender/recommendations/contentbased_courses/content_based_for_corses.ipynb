```python
import torch
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import ast
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("../DataSets/processed/courses_content_based_data_after_eda_after_embedding.csv",sep=";")
```

```python
df.head()
```

Load model

```python
model = SentenceTransformer("all-MiniLM-L6-v2")  
```

```python
df.head()
```

```python
# df.columns = ["id","course_name", "course_description","degree_description","degree_name","stream_name", "similarity"]
```

```python
def recommend_courses(input_text, top_n=5):
    input_embedding = model.encode([input_text])

    df["similarity"] = df["embeddings"].apply(lambda x: cosine_similarity([x], input_embedding)[0][0])

    recommendations = df.sort_values(by="similarity", ascending=False).head(top_n)

    return recommendations  
```

```python
index = eval(input())
input_text = df["combined_text"].iloc[index]
print(f"User Searched for : {df["course_names"].iloc[index]}")
recommended_courses = recommend_courses(input_text)
recommended_courses
```
